We want to do something here. As described in Regression Notebook 4 (MLR Predictive Model Selection) we can decide what subset of these attributes most effectively determines release year data. The next cell uses much code adapted from that notebook

```python
#maybe a larger PCA will help. Look at the next piece of data.
#Actually PCA use may have been unnecessary as a whole. 2 things to try: polynomial dataset (which helps a bit but I need more columns), individualized graph examination, and possibly neural-net. But I think a neural network here would be overkill.

#Hey, simpler is often better. Let's try a basis MLR Prediction. SVM might be too complicated with PCA as all the data modification makes it downright useless

#We'll copy these functions from the MLR notebook. This isn't the final result but it will allow us to narrow down attributes
def get_X_y(df,features,target):
    # Returns X then y
    return np.array(df[features]), np.array(df[target])
def get_abce(model, X, y):
    # get the prediction
    pred = model.predict(X)
    
    # Returns the error as sum of absolute value
    return np.sum(abs(pred-y))/len(y)
def powerset_no_empty(s):
    power_set = []
    x = len(s)
    for i in range(1 << x):
        power_set.append([s[j] for j in range(x) if (i & (1 << j))])
            
    return power_set[1:]

possible_features = powerset_no_empty(['danceability','energy','loudness','speechiness','acousticness','liveness','valence','duration_ms'])
from sklearn.linear_model import LinearRegression
reg64=LinearRegression(copy_X = True)
errors = np.empty((5,len(possible_features)))
baseerrors=np.empty((5,len(possible_features))) #these are raw errors, with sign; checks for bias, which does not appear much
                                       
i = 0

## This is for the initial input into the kfold object
XX,yy= get_X_y(mcore9, possible_features[0], 'release_year')

## Perform cross validation
for train_index, test_index in kfold.split(XX,yy):
    ## For each possible model
    for j in range(len(possible_features)):
        ## get X and y
        XX, yy = get_X_y(mcore9, possible_features[j], 'release_year')


        # Get the cv train test split
        X_train, X_test = XX[train_index], XX[test_index]
        y_train, y_test = yy[train_index], yy[test_index]
        
        # Cloning the regression makes a fresh regression 
        # model for each run
        clone_reg64 = clone(reg64)
        
        # fit the model
        clone_reg64.fit(X_train,y_train)
        
        errors[i,j] = get_abce(clone_reg64, X_test, y_test) #absolute value error
        baseerrors[i,j] = np.sum((clone_reg64.predict(X_test)-y_test))/len(y_test)
    ## We'll now move to the next split
    i = i + 1
e=errors.mean(axis=0)
b=baseerrors.mean(axis=0)
np.argmin(e)

print("The model with the lowest mean CV error",
     "was the one with", possible_features[np.argmin(np.mean(errors, axis = 0))],
     "as the features. This model had a mean CV error of",
     np.round(np.min(np.mean(errors, axis=0)),5))

#this concurs with what we said about liveness changing little over time, duration also doesn't matter much because it still varies a lot particualrly among genres
e,b

#Checking b, there's almost no bias, all variance. So at least it's good on average!
```
